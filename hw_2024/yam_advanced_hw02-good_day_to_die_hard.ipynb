### Задача 1: доверительные интервалы для долей (25 баллов)

В этой задаче вам предстоит поработать с разными видами доверительных интервалов для долей.

Пусть случайные величины $X_1, \ldots X_n$ независимы и имеют распределение Бернулли с неизвестным параметром $p$, при этом $\hat p = \bar x$. 

__а) [5 баллов]__ Если воспользоваться ЦПТ, можно построить для неизвестного $p$ доверительный интервал Вальда. На лекции мы для этого использовали сходимость 
$$
\frac{\hat p - p}{\sqrt{\frac{\hat p (1 - \hat p)}{n}}} \to N(0;1),
$$

а затем решали неравенство 
$$
-z_{cr} \leq \frac{\hat p - p}{\sqrt{\frac{\hat p (1 - \hat p)}{n}}} \leq z_{cr}.
$$

Получалось, что при уровне значимости $\alpha$ доверительный интервал 

$$
\hat p \pm z_{1 - \frac{\alpha}{2}} \cdot \sqrt{ \frac{\hat p \cdot (1 - \hat p)}{n} }
$$

накрывает неизвестное значение $p$ с вероятностью $1 - \alpha$. Мы в этом задании с помощью симуляций на компьютере сравним фактическую вероятность накрытия неизвестного параметра $p$ интервалами Вальда, Вильсона и Агрести—Коуллаc номинальной 95\%-й вероятностью.

Одна из ошибок заключается в некорректном бейзлайне для сравнения своего текста. Авторы решили использовать тест Стьюдента, позабыв, что в его предпосылках сказано, что выборка должна быть нормальной, а дисперсии выборок должны совпадать. Вам в этом задании предлагается самостоятельно разработать тест для проверки гипотез в рамках предпосылки о лог-нормальности выборок. 

Разработку теста мы сделаем двумя способами, а затем удостоверимся, что в АА-тестировании тест не разваливается. 

```python
from collections import defaultdict

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [8, 6]

def plot_pvalue_distribution(dict_pvalues):
    """
        Рисует графики распределения p-value.
    """
    X = np.linspace(0, 1, 1000)

    for key, pvalues in dict_pvalues.items():
        Y = [np.mean(pvalues < x) for x in X]
        plt.plot(X, Y, label=key)

    plt.plot([0, 1], [0, 1], '--k', alpha=0.8)
    plt.title('Оценка распределения p-value', size=16)
    plt.xlabel('p-value', size=12)
    plt.legend(fontsize=12)
    plt.grid()
    plt.show()
```

```python
sample_size = 10
sigma = 2
dict_pvalues = defaultdict(list)

for _ in range(10000):
    # выборки генерируется из лог-нормального распределения с одинаковыми параметрами 
    a, b = np.random.lognormal(sigma=sigma, size=(2, sample_size))

    # тестируем гипотезу t-тестом, она должна не отвергаться в 95% случаев 
    _, pvalue = stats.ttest_ind(a, b)
    dict_pvalues['A/A'].append(pvalue)

# отрисовываем для АА-теста рок-кривую и видим, что всё развалилось
plot_pvalue_distribution(dict_pvalues)
```

Начнём с теста отношения правдоподобий. Говорят, что случайная величина имеет логнормальное распределение  

$$X \sim \mathrm {LogN} (\mu ,\sigma ^{2}),$$

если она задаётся плотностью распределения

$$
{\displaystyle f_{X}(x)={\frac {1}{x\sigma {\sqrt {2\pi }}}}e^{-(\ln x-\mu )^{2}/2\sigma ^{2}}.}
$$

Логарифм такой случайной величины будет иметь нормальное распределение $\ln X \sim N(\mu, \sigma^2),$  при этом $\mathbb{E}(X) = e^{\mu + \frac{\sigma^2}{2}}.$ 

> Цены на разную продукцию часто имеют распределение похожее на логнормальное. Довольно часть в ценах есть выбросы и хвост распределения оказывается тяжелее нормального. Логарифмирование можно обобщить до процедуры Бокса-Кокса и сгладить распределение до нормального, но это работает не всегда.
